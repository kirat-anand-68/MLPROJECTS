```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
image_as_array=mpimg.imread('sunset-palm-silhouette-stockcake.webp')
```

```python
image_as_array.shape#(H,W,Color channels)
```

```python
#plt.figure(dpi=200)
plt.imshow(image_as_array)
```

```python
#pli imshow needs the image as aray not real image as jpeg 
```

```python
(h,w,c)=image_as_array.shape
```

```python
c
```

```python
image_as_array2d=image_as_array.reshape(h*w,c)
```

```python
len(image_as_array.shape)
```

```python
len(image_as_array2d.shape)
```

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(n_clusters=6)
```

```python
labels=model.fit_predict(image_as_array2d)
```

```python
labels
```

```python
rgb_codes=model.cluster_centers_.round(0).astype(int)
```

```python
rgb_codes
```

```python
import numpy as np
```

```python
quantized_image=np.reshape(rgb_codes[labels],(h,w,c))
```

```python
plt.imshow(quantized_image)
```
