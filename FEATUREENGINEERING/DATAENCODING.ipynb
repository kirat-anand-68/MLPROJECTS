```python
#NOMINAL OR OHE Encoding
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.DataFrame({
'color':['red','blue','green','green','red','blue']
})
```

```python
df.head()
```

```python
#create an instance of onehot encoder
encoder=OneHotEncoder()
```

```python
encoded=encoder.fit_transform(df[['color']]).toarray()
```

```python
encoder_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
```

```python
encoder_df
```

```python
encoder.transform([['blue']]).toarray()
```

```python
#ordinal Encoding
# encoded the categorical data
from sklearn.preprocessing import OrdinalEncoder


```

```python
df=pd.DataFrame({
    'size':['small','medium','large','medium','small','large']
})
```

```python
df
```

```python
## create an instance of OrdinalEncoder and tehn fit_transform
```

```python
encoder=OrdinalEncoder(categories=[['small','medium','large']])
```

```python
encoder.fit_transform(df[['size']])
```

```python
encoder.transform([['small']])
```

```python
#target guided ordinal encoding technique
df=pd.DataFrame({
    'city':['New York','London','Paris','Tokyo','New York','paris'],
    'price':[200,150,300,250,180,320]
})
```

```python
df.head()
```

```python
mean_price=df.groupby('city')['price'].mean().to_dict()
```

```python
mean_price
```

```python
df['city_encoded']=df['city'].map(mean_price)
```

```python
df[['city','city_encoded']]
```
